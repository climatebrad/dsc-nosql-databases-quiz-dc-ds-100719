# NoSQL Databases - Quiz


## Introduction

In this lesson, you'll answer some common open-ended interview questions about NoSQL databases.


## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
A SQL database contains tables with fixed indexes and relations between them, supporting SQL queries. There are a variety of NoSQL databases, meant for managing types of data, relationships, and amount of data that SQL databases are unsuited for. 




## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
One example is the nodes and connections in a social media network; the number of 3rd-degree neighbors of a node would be explosively difficult for a SQL database to return. In this case, a graph database should be used.  





## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
MongoDB stores "documents", which are hash tables (ala python dict or JSON format) which can have key-value pairs of any kind. The number and names of the keys and the types of values is not predefined at table creation.








## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
Fault tolerance describes the ability of a system to continue operating even when there is component failure. Resilient Distributed Datasets are distributed in components on multiple servers, with every server serving as backups to the others. The RDD will continue to function even if individual servers go down.








## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
MapReduce describes the process of separating a database query into two components - direct information retrieval, which can be done in parallel on dataset components (map), and aggregation of the results (reduce). Hadoop is one of the most popular MapReduce database engines.
